## sales analysis 

use line plot to track monthly sales and histograms for order quantity distibution
customer demographics: visualize age distribution(histogram) and spending patterns(scatter plot).

compare student marks across subjects(bar chart) and identify outliers(box plot)

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\teju\\Desktop\\r\\E-commerce Customer Behavior - Sheet1.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Satisfaction Level'].value_counts()
```

```python
df['Satisfaction Level'].fillna('Satisfied')
```

```python
df['Satisfaction Level'].value_counts()
```

```python
df['Month'] = pd.cut(
    df['Days Since Last Purchase'], 
    bins=[0, 10, 20, 30, 40, 50, 60],
    labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
)
monthly_sales = df.groupby('Month', observed=False)['Total Spend'].sum()

plt.figure(figsize=(10, 6))
monthly_sales.plot(marker='X', color='green')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
#histogram for items purchased
plt.hist(df['Items Purchased'], bins=10, color='lavender', edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Items Purchased')
plt.ylabel('Frequency')
plt.show()
```

# Age distribution Histogram

```python
plt.hist(df['Age'],bins=10,color='pink',edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['Items Purchased'],df['Total Spend'],color='skyblue',marker='*',edgecolor='black',s=500)
plt.xlabel('Items Purchased')
plt.ylabel('Total spend')
plt.title('Spending Patterns')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# school performance:
## compare student marks across subjects(bar chart) and identify outliers(box plot)

```python
hd=pd.read_csv("C:\\Users\\teju\\Desktop\\r\\study_performance.csv")
hd.head()
```

```python
subject_means = hd[['math_score', 'reading_score', 'writing_score']].mean()
plt.bar(subject_means.index, subject_means.values, color=['skyblue', 'lavender', 'pink'], edgecolor='black', linewidth=2)
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.title('Average Marks Across Subjects')
plt.show()
```

```python
hd[['math_score', 'reading_score', 'writing_score']].boxplot()
plt.title('Box Plot of Scores Across Subjects')
plt.ylabel('Marks')
plt.xlabel('Subjects')
plt.show()
```
