```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\tejas\\OneDrive\\Desktop\\54\\GlobalLandTemperaturesByCity.csv")
df.head()
```

```python
city_name = "Århus" 
filtered_data = df[df['City'] == city_name][['dt', 'AverageTemperature']]
filtered_data['dt'] = pd.to_datetime(filtered_data['dt'])
filtered_data['Year'] = filtered_data['dt'].dt.year
filtered_data = filtered_data.dropna(subset=['AverageTemperature'])
annual_avg_temp = filtered_data.groupby('Year')['AverageTemperature'].mean().reset_index()
print("Annual Average Temperatures:")
print(annual_avg_temp)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(annual_avg_temp['Year'], annual_avg_temp['AverageTemperature'], color='blue', label='Avg Temperature')
plt.plot(annual_avg_temp['Year'], annual_avg_temp['AverageTemperature'], color='red')
plt.title(f"Average Annual Temperature Trend for {city_name}", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Temperature (°C)", fontsize=12)
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
