```python
import re

def clean_text(text):
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

data['JobDescription'] = data['JobDescription'].apply(clean_text)
data['JobRequirment'] = data['JobRequirment'].apply(clean_text)
```

```python
from sklearn.preprocessing import LabelEncoder
data1 = {'Salary': ['Competitive', 'Salary is dependent on education, experience, and\r\ncapabilities, and will be discussed on a case-by-case basis.', 'Negotiable', '800,000 AMD - 1,000,000 AMD (Gross) plus bonuses.', 'Highly competitive plus a social package and\r\nopportunity for training abroad.','Based on the companys remuneration scheme.'],
        'Test Score': [50000, 52000, 54000, 56000, 58000,60000]}
df = pd.DataFrame(data1)
encoder = LabelEncoder()
df['Salary_range'] = encoder.fit_transform(df['Salary'])
df
```

```python
import re

def extract_salary_range(salary_str):
    pattern = r'(\d+)(k?)-(\d+)(k?)\s*(\w+)?'
    match = re.search(pattern, salary_str)
    if match:
        min_salary = int(match.group(1)) * 1000 if match.group(2) == 'k' else int(match.group(1))
        max_salary = int(match.group(3)) * 1000 if match.group(4) == 'k' else int(match.group(3))
        return min_salary, max_salary
    else:
        return None, None
        
data[['min_salary', 'max_salary']] = data['Salary'].apply(lambda x: pd.Series(extract_salary_range(str(x))))

data[['Salary', 'min_salary', 'max_salary']]
```

```python
def categorize_education(eligibility):
    eligibility = str(eligibility).lower()
    
    if 'bachelor' in eligibility or 'undergraduate' in eligibility:
        return 'Bachelor\'s Degree'
    elif 'master' in eligibility or 'graduate' in eligibility:
        return 'Master\'s Degree'
    elif 'phd' in eligibility or 'doctorate' in eligibility:
        return 'PhD'
    elif 'high school' in eligibility or 'secondary' in eligibility:
        return 'High School'
    elif 'associate' in eligibility:
        return 'Associate Degree'
    else:
        return 'Not specified'

def categorize_experience(eligibility):
    eligibility = str(eligibility).lower()
    
    if 'fresher' in eligibility or 'entry level' in eligibility:
        return '0-1 years'
    elif '1-2 years' in eligibility or '1 year' in eligibility:
        return '1-2 years'
    elif '2-5 years' in eligibility or '3-5 years' in eligibility:
        return '2-5 years'
    elif '5-10 years' in eligibility:
        return '5-10 years'
    elif '10+ years' in eligibility:
        return '10+ years'
    else:
        return 'Not specified'
        
def categorize_audience(audience):
    audience = str(audience).lower()
    
    if 'all interested' in audience or 'all candidates' in audience:
        return 'Open to all'
    elif 'qualified candidates' in audience or 'professionals' in audience:
        return 'Qualified professionals'
    elif 'specific professionals' in audience:
        return 'Specific professionals'
    else:
        return 'Not specified'

data['Education_Level'] = data['Eligibility'].apply(categorize_education)
data['Experience_Range'] = data['Eligibility'].apply(categorize_experience)
data['Audience_Category'] = data['Audience'].apply(categorize_audience)

data[['Eligibility', 'Education_Level', 'Experience_Range', 'Audience', 'Audience_Category']]
```
